## ラベルをテキストから数値(整数)に変換するコード
データ解析において、文字データをそのまま扱うことが難しい場合があります。
例えば、回帰問題や分類問題など…

その場合、このコードでテキストを数値に変換しましょう。

---
ここでは例として、`car_price.csv`のラベルを変換してみます。
以下のサイトからcsvアイルを入手してください。

car_price.csv \\
https://github.com/YosukeSugiura/datamining-excercise/blob/master/5_excercise/car_price.csv


> 参考まで、下記はオリジナルデータセットを提供する Kaggle のサイトです。
>
>car_price dataset (original) \\
>https://www.kaggle.com/datasets/ngawangchoeda/car-price-dataset

```python
#
#   Transform Class Name to Class Value
#
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# データ読み込み
file_name = 'car_price.csv'
data = pd.read_csv(file_name)
keys = data.keys()      # キー(項目)の名前を取得

# データのコピー
data_new = data.copy()

# pandas形式からリスト形式に変換
att  = keys[2]          # 変換したいキー(項目)を指定
data_pds = data[att]    # extract a column
data_lst = list(data_pds.values)

# データの確認
le = LabelEncoder()     # Label Encoder Obj
le.fit(data_lst) 
print('\nTotal {0} types.'.format(len(le.classes_)))

# テキストから数値に変換
data01 = le.transform(data_lst)

# 表示
print('[ ID : `{0}` ]'.format(att))
for i, name in enumerate(le.classes_):
  print(' - {0}: {1}'.format(i, name))

# データの上書き
data_new[att] = data01

# データの保存
data_new.to_csv(file_name, index=False)
print('Complete Overwriting')
```
